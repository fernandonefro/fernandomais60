```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
ativos = ['PETR4.SA','VALE3.SA','WEGE3','RADL3.SA','OIBR3.SA','KNRI11.SA','SMAL11.SA','AAPL34.SA','IVVB11.SA','IVVB11.SA','ETH-USD','USDBRL=X','CASH3.SA']
```

```python
inicio = '2020-05-01'
fim = '2021-11-03'
```

```python
precos = pd.DataFrame()
for i in ativos:
  precos[i] = yf.download(i, start = inicio, end = fim)['Adj Close']
```

```python
precos['ETH-BRL'] = precos['ETH-USD']*precos['USDBRL=X']
```

```python
precos = precos.drop(columns=['ETH-USD','USDBRL=X'])
```

```python
normalizado = precos/precos.iloc[0]
```

```python
normalizado.plot(figsize = (8,8));
```
